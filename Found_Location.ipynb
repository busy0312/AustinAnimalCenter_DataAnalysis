```python
import pandas as pd
```

```python
# Read csv file_intake="raw_data/Austin_Animal_Center_Intakes.csv"
found_location="Austin_Animal_Center_Found_Pets_Map_coordinates.csv"
```

```python
found_location_map=pd.read_csv(found_location)
found_location_map.head()
```

```python
found_location_map = found_location_map.join(found_location_map["Found Location"].str.split("(", expand=True).add_prefix("Coord"))
found_location_map
```

```python
found_location_map = found_location_map.join(found_location_map["Coord1"].str.split(",", expand=True).add_prefix("Location"))
found_location_map
```

```python
found_location_map["Location1"]=found_location_map["Location1"].str.replace(")", "")
found_location_map
```

```python
found_location_map["Coord1"]=found_location_map["Coord1"].str.replace(")", "")
found_location_map
```

```python
found_location_map.to_json("Found_Location.json", orient='records')
```
